```python
import requests
```

```python
uid='2fb35543-5ca7-47fb-b7e8-f3f071869dc5'
username='samar'
pin='123456'
base_url = 'http://localhost:8080'
```

```python
signup_url = '/auth/register'

resp = requests.post(base_url + signup_url, json={
    'username': 'samar',
    'password': '12345678',
    'pin': '123456',
    'email': 'samar@gmail.com',
    'user_type': 'business',
    'is_admin': True
})

print(resp.json())
```

```python
resp = requests.post(f'{base_url}/auth/login', json={
    'username': 'samar',
    'pin': '123456'
})
print(resp.status_code)
print(resp.json())
token = resp.json().get('token')
```

```python
resp = requests.get(f'{base_url}/auth/user/info', headers={'Authorization': f'Bearer {token}'})
print(resp.status_code)
print(resp.json())
```

```python
data=resp.json()
DriveID = data['drives'][0]['DriveID']
```

```python
invite = '/auth/drive/invite'
resp = requests.post(f'{base_url}{invite}', json={'drive_id': DriveID, 'access': 'read'}, headers={'Authorization': f'Bearer {token}'})
print(resp.status_code)
print(resp.json())
```

```python
invite_code = resp.json().get('invite_code')
######### new user #########
```
